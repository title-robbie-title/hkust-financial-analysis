```python
import pandas as pd
import matplotlib.pyplot as plt 
fb = pd.read_csv('Data/facebook.csv', index_col='Date')
fb['price_1'] = fb['Close'].shift(-1)
fb['PriceDiff'] = fb['price_1'] - fb['Close']
fb['returns'] = fb['PriceDiff'] / fb['Close']
fb['direction'] = [1 if fb.loc[ei, 'PriceDiff'] > 0 else -1 for ei in fb.index]
fb['average3'] = (fb['Close'] + fb['Close'].shift(1) +fb['Close'].shift(2)) /3
fb['MA40'] = fb['Close'].rolling(40).mean()
fb['MA200'] = fb['Close'].rolling(200).mean()
fb['Close'].plot()
fb['average3'].plot()
plt.show()
fb.head()
#creating a plot to compare closing price and 3-day moving average
#not enough data points for MA200 to be plotted
#some traders beleive that when the short-term moving average crosses above the long-term moving average, it signals a buying opportunity, and vice versa for a selling opportunity
```
