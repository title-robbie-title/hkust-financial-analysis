```python
import pandas as pd
import matplotlib.pyplot as plt
fb = pd.read_csv('Data/facebook.csv')
fb['MA10'] = fb['Close'].rolling(10).mean() # fast signal
fb['MA50'] = fb['Close'].rolling(50).mean() # slow signal
fb['Close1'] = fb['Close'].shift(-1)
fb['Shares'] = [1 if fb.loc[ei, 'MA10'] > fb.loc[ei, 'MA50'] else 0 for ei in fb.index]
fb['Profit'] = [fb.loc[ei, 'Close1'] - fb.loc[ei, 'Close']
    if fb.loc[ei, 'Shares'] == 1 else 0 for ei in fb.index] 
fb['wealth'] = fb['Profit'].cumsum()
fb.head()
fb.tail()
#calculating cumulative wealth based on moving average trading strategy
# Calculating cumulative wealth based on moving average trading strategy
# MA10/MA50 crossover strategy:
# - Buys when MA10 crosses above MA50
# - Stays out when MA10 <= MA50
#
# In this dataset, MA50 dominates toward the end, so the strategy signals "no buy" for the final period.
# The cumulative wealth shows that the strategy made a total profit earlier (â‰ˆ 556.84 units),
# but finished flat near the end because no recent buy signals occurred.


```

```python
import pandas as pd
import matplotlib.pyplot as plt
fb = pd.read_csv('Data/facebook.csv')
fb['MA10'] = fb['Close'].rolling(10).mean() # fast signal
fb['MA50'] = fb['Close'].rolling(50).mean() # slow signal
fb['Close1'] = fb['Close'].shift(-1)
fb['Shares'] = [1 if fb.loc[ei, 'MA10'] > fb.loc[ei, 'MA50'] else 0 for ei in fb.index]
fb['Profit'] = [fb.loc[ei, 'Close1'] - fb.loc[ei, 'Close']
    if fb.loc[ei, 'Shares'] == 1 else 0 for ei in fb.index] 
fb['wealth'] = fb['Profit'].cumsum()
print("Total money you win is ", fb.loc[fb.index[-2], 'wealth'])
print("Total money you spent is ", fb.loc[fb.index[0], 'Close'])
fb['wealth'].plot()
#printing total profit and initial investment based on moving average trading strategy
```
