```python
import pandas as pd
import matplotlib.pyplot as plt

ms = pd.read_csv('Data/microsoft.csv')
ms['MA10'] = ms['Close'].rolling(10).mean()
ms['MA50'] = ms['Close'].rolling(50).mean()
plt.figure(figsize=(12,6))
plt.plot(ms['Close'], label='MSFT Close')
plt.plot(ms['MA10'], label='MSFT MA10')
plt.plot(ms['MA50'], label='MSFT MA50')
plt.title('Microsoft Trading Strategy â€“ MA10 & MA50')
plt.legend()
plt.show()
#visualizing microsoft stock price with moving averages


```
