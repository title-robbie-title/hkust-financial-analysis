```python
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Apple data
aapl = pd.read_csv("data/apple.csv", index_col='Date')

# Compute log returns
aapl['LogReturn'] = np.log(aapl['Close']).shift(-1) - np.log(aapl['Close'])

# Mean and std of Apple returns
mu = aapl['LogReturn'].mean()
sigma = aapl['LogReturn'].std()

# Build x-values around typical daily returns (-10% to +10%)
x = np.arange(-0.1, 0.1, 0.0001)
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF curve
plt.ylim(0, 20)
plt.plot(x, pdf)

# Shaded region (loss between -10% and -1%)
shade_x = np.arange(-0.1, -0.01, 0.0001)
shade_y = norm.pdf(shade_x, mu, sigma)

plt.fill_between(shade_x, shade_y, 0, facecolor='pink', alpha=0.5)

plt.title("Shaded Probability Region on Normal Curve")
plt.xlabel("Return")
plt.ylabel("PDF")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Apple data
aapl = pd.read_csv("data/apple.csv", index_col='Date')

# Compute log returns
aapl['LogReturn'] = np.log(aapl['Close']).shift(-1) - np.log(aapl['Close'])

# Mean and std of Apple returns
mu = aapl['LogReturn'].mean()
sigma = aapl['LogReturn'].std()

# Build x-values around typical daily returns (-10% to +10%)
x = np.arange(-0.1, 0.1, 0.0001)
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF curve
plt.ylim(0, 20)
plt.plot(x, pdf)

# Shaded region (loss between -10% and -1%)
shade_x = np.arange(-0.1, -0.01, 0.0001)
shade_y = norm.pdf(shade_x, mu, sigma)

plt.fill_between(shade_x, shade_y, 0, facecolor='pink', alpha=0.5)

plt.title("Shaded Probability Region on Normal Curve")
plt.xlabel("Return")
plt.ylabel("PDF")
plt.grid(True)
plt.show()
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The probability is', prob_return1)
# Annualized mean and standard deviation over 220 days
mu220 = 220 * mu
sigma220 = (220 ** 0.5) * sigma


print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))
#this code calculates and visualizes the probability of Apple's stock experiencing a loss greater than 40% in a year,m based on the annualized normal distribution of its daily log returns


```
